```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score
from scipy.stats import pearsonr, spearmanr
```

```python
pearson_df = pd.DataFrame(columns=['Category','Omics','Corr'])
pearson_df.head()

spearman_df = pd.DataFrame(columns=['Category','Omics','Corr'])
spearman_df.head()
```

```python
pred_df = pd.read_table("/cellar/users/kivil/Data/Projects/DRPredICT/fit/tetraomic/multiomic/version_1/outputs.tsv")
pred_df = pred_df.rename(columns={"outputs": "pred_aucs", "aucs": "true_aucs"})
tetra_df = pred_df.copy()
print(pred_df.shape)
pred_df.head()
```

```python
cat = 'Tetraomic'
omics = 'mut+exp+cn+drug'

g = sns.scatterplot(data=pred_df, x="pred_aucs", y="true_aucs")
g.plot([-0.2, 2], [-0.2, 2], ls="--", c=".3", alpha=0.5)
g.set_title("%s (%s)" % (cat, omics), fontsize=14)

p_corr = pearsonr(pred_df["pred_aucs"], pred_df["true_aucs"])[0]
s_corr = spearmanr(pred_df["pred_aucs"], pred_df["true_aucs"])[0]

pearson_df = pearson_df.append({'Category':cat, 'Omics':omics, 'Corr':p_corr}, ignore_index=True)
spearman_df = spearman_df.append({'Category':cat, 'Omics':omics, 'Corr':s_corr}, ignore_index=True)

g.annotate("pearsonr = %s" % str(p_corr)[:5], (-0.2, 1.6), size=12)
g.annotate("spearmanr = %s" % str(s_corr)[:5], (-0.2, 1.8), size=12)
```
