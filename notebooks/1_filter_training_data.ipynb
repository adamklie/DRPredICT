```python
import pandas as pd
import numpy as np
from scipy import stats

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_ext = "/cellar/users/kivil/Data/Projects/hackathons/data/multiomic_drug_response/training"
```

```python
exp_df = pd.read_table("%s_expression.tsv" % (file_ext), index_col=0)
print(exp_df.shape)
exp_df.head()
```

```python
top2000_genes = sorted(exp_df.var().sort_values(ascending=False)[:2000].index)
print(len(top2000_genes), top2000_genes[:5])
```

```python
exp_df = exp_df[top2000_genes]
exp_df
```

```python
exp_df.to_csv("training_top2000_raw_expression.tsv", sep='\t', index=True)
```

```python
stats.zscore(exp_df, axis=0)
```

```python
exp_df_norm = pd.DataFrame(stats.zscore(exp_df, axis=0), index=exp_df.index, columns=exp_df.columns)
exp_df_norm
```

```python
np.mean(exp_df_norm)
```

```python
np.std(exp_df_norm)
```

```python
exp_df_norm.to_csv("training_top2000_expression.tsv", sep='\t', index=True)
```
